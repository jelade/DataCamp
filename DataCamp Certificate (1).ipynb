```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("travel_insurance.csv")
```

```python
df.head()
```

```python
df['FamilyMembers'].unique()
```

```python
df1 = df.groupby('TravelInsurance')['TravelInsurance'].count()
df2 = df.groupby('EverTravelledAbroad')['EverTravelledAbroad'].count()
df2
```

```python
bar=['No','Yes ']
df1.plot(kind='pie', labels = bar,autopct='%1.1f%%')
#plt.xticks(range(2),bar, rotation = 0)
#plt.xlabel('Travel Insurance')
plt.title('Travel Insuance')
plt.show()
plt.savefig('Travel Insuance.png')
```

```python
bar=['No','Yes']
df2.plot(kind='pie', labels = bar,autopct='%1.1f%%')
#plt.xticks(rotation = 0)
#plt.xlabel('Ever Travelled Abroad')
plt.title('Travel habit')
plt.show()
plt.savefig('Travel habit.png')
```

```python
df2 = df.groupby('EverTravelledAbroad')['EverTravelledAbroad'].count()
df2
```

```python
sns.catplot(x='EverTravelledAbroad', hue='TravelInsurance',
                data=df, saturation=.5,
                kind="count", ci=None, aspect=.6)

plt.title('Travel habit vs Travel Insurance')
plt.show()
plt.savefig('Travel habitvsinsurance.png')
```

```python
bar=['No','Yes']
sns.catplot(x='TravelInsurance', hue= 'EverTravelledAbroad',
                data=df, saturation=.5,
                kind="count", ci=None, aspect=.6)
plt.xticks(range(2),bar)
plt.title('Travel habit vs Travel Insurance')
plt.show()
plt.savefig('Travel habitvsinsurance2.png')
```

```python
bar=['No','Yes']
sns.catplot(x='TravelInsurance', hue = 'Age',
                data=df, saturation=1.5,
                kind="count", ci=None, aspect=.6)
plt.xticks(range(2),bar)
plt.title('Age vs Travel Insurance')
plt.show()
plt.savefig('Agevsinsurance.png')
```

```python
g = sns.catplot(x = 'TravelInsurance', y = 'AnnualIncome', data = df,  estimator = np.mean, kind = 'bar')

(g.set_axis_labels("Travel Insurance", "Average Annualincome")
  .set_xticklabels(['No','Yes']))

plt.title('Annual Income vs Travel Insurance')
plt.show()
```
